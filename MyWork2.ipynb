```python
import pandas as pd
import matplotlib.pyplot as plt
import wordcloud as wc
```

```python
df = pd.read_csv('24-06-2021.csv')
```

```python
df
```

```python
#Combined all row values in cleaned_definition
cleaned_definition_str = ' '.join(df['cleaned_definition'])
cleaned_definition_str
```

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(max_words=50, background_color= 'white').generate(cleaned_definition_str)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
#Split the df into two df based on the value 0 or 1 for misogyny and not so

df_nonmisogynist = df[df['is_misogyny']==0.0]

df_nonmisogynist
```

```python
df_misogynist = df[df['is_misogyny']==1.0]
df_misogynist
```

```python
# Repeat step above to combine definition into a long string for each df
nm_cleaned_definition_str = ' '.join(df_nonmisogynist['cleaned_definition'])
m_cleaned_dfinition_str = ' '.join(df_misogynist['cleaned_definition'])
```

```python
#Generate wordcloud for both. 1st one is non-misogynist
nm_wc = WordCloud(max_words=50, background_color= 'white').generate(nm_cleaned_definition_str)
plt.imshow(nm_wc)
plt.axis('off')
plt.show()
```

```python
#2nd word cloud for misogynist 
m_wc = WordCloud(max_words =50, background_color ='white').generate(m_cleaned_dfinition_str)
plt.imshow(m_wc)
plt.axis('off')
plt.show()
```
